```python
# 연속된 0의 개수 list화해서 max 뽑아내기
# type을 항상 조심하자


myinput = input('숫자를 입력하세요')
countlist = []
count = 0

for i in range(len(myinput)):
  if myinput[i] == '0':
    count += 1
# 0이 아닌 애를 만나면 count를 초기화해라
  else : 
    if count > 0: # 0이 자꾸 저장되면 메모리 소모가 심하므로 0을 제외한 count만 저장되도록
      countlist.append(count)
      count = 0 
countlist.append(count)

print(countlist)
max(countlist)
```

```python
def count_max_0():
  myinput = input('숫자를 입력하세요')
  countlist = []
  count = 0
  for i in range(len(myinput)):
    if myinput[i] == '0':
      count += 1
      
    else : 
      if count > 0:
        countlist.append(count)
        count = 0
  countlist.append(count)
  return max(countlist)

count_max_0()
```

```python
# __init__(self) : -> class를 호출할 때 가장 먼저 실행이된다/ 생성자
# main 특성(빠질 수가 없는 특성)을 주로 넣는다
```

```python
class calc:
  
  def __init__(self,input_a,input_b):
    self.input_a = input_a
    self.input_b = input_b
    
  def add(self):
    return self.input_a + self.input_b
  def sub(self):
    return  self.input_a - self.input_b
  def mul(self):
    return  self.input_a * self.input_b
  def div(self):
    return  self.input_a / self.input_b


이진재 = calc(4,5)
print(이진재.add())
```

```python
class supermarket:
  count = 0
  
  def __init__(self,location : str,name : str,product : list,customer : int):
    self.location = location
    self.name = name
    self.product = product
    self.customer = customer
    supermarket.count += 1
    
  def print_location(self):
    print(self.location)
    
  def change_category(self,x : list) :
    self.product = x
  
  def show_list(self) :
    print(self.product)
    
  def enter_customer(self):
    self.customer += 1
    
  def show_info(self):
    print(f'{self.name} ,{self.location},{self.product},{self.customer}')
    
  def show_supermarket_count():
    print(supermarket.count)
    
Super_supermarket = supermarket("왕십리","Super",["포카칩","사이다","맥주"],3)
Everything_supermarket = supermarket("잠실","Everything",["통크","소보루빵","돼지바","제로콜라"],6)

Super_supermarket.print_location()
Everything_supermarket.enter_customer()
Everything_supermarket.change_category(["김밥","팝콘"])
print(Everything_supermarket.product)
print(supermarket.count)
Everything_supermarket.show_info()
```
